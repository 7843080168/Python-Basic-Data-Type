```python
# Example - 7
# Subplots
# Create subplots in trignometric function
# compute the x and y coordinates for points on sine and cosine curves
x = np.arange(0,5 * np.pi,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)
```

```python
# Set up a subplot grid that has height 2 and width 1, 
# and set the first such subplot as active.
plt.subplot(2,1,1)
# Make the first plot
plt.plot(x,y_sin,'r--')
plt.title('Sine')
```

```python
# Set the second subplot as active, and make the second plot. 
plt.subplot(2, 1, 2) 
plt.plot(x, y_cos,'g--') 
plt.title('Cosine')   
# Show the figure. 
plt.show()
```

```python
# ####OR ONE CANVAS IN SUBPLOT#### #
#Subplot()
# Compute the x and y coordinates for points on sine and cosine curves 
x = np.arange(0, 5 * np.pi, 0.1) 
y_sin = np.sin(x) 
y_cos = np.cos(x)  
   
# Set up a subplot grid that has height 2 and width 1, 
# and set the first such subplot as active. 
plt.subplot(2, 1, 1)
   
# Make the first plot 
plt.plot(x, y_sin,'r--') 
plt.title('Sine')  
   
# Set the second subplot as active, and make the second plot. 
plt.subplot(2, 1, 2) 
plt.plot(x, y_cos,'g--') 
plt.title('Cosine')  
   
# Show the figure. 
plt.show()
```

```python
# Data to plot
labels = 'python','C++','Ruby','Java'
sizes = [215,130,245,210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.4, 0, 0, 0) ## explode 1st slice
#plot
plt.pie(sizes,explode=explode,labels=labels,colors=colors,
autopct='%1.1f%%',shadow=False)
plt.axis('equal')
plt.show()
```
